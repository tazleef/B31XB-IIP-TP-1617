Blend the lena image and vibot image using simple, alpha and pyramid method.   

Pyramid blending method: 
1. Create a mask the same size as your input images 
2. Create Laplacian pyramid of the input images 
    2.1. Create Gaussian pyramids of the input images
    2.2. Create Laplacian pyramids using the Gaussian pyramids
3. Create Gaussian pyramid for the region mask
4. Blend the two laplacian pyramids using the pyramid of the mask as weigh 
    $$ BL = L_1 * (GM) + L_2 *(1 - GM)$$
5. Colapse the pyramid to get the final image 
    $$rescale(rescale(rescale(BL_3)+BL_2)+BL_1)+BL_0$$

Note : For pyramid blending approach, you can use:
    * transform.rescale for upscaling your images
    * transform.pyramid_gaussian for building your pyramids
    * The output of transform.pyramid_gaussian is a type generator and you need to convert it to a list :
    
        b = [a for a in Gaussian_pyramid]


```python
##Defining simple blending
def simple(image1,image2):
    image1 = img_as_float(image1)
    image2 = img_as_float(image2)
    
    image_out = np.zeros(image1.shape) ##making a third image to store the final blended image
    ##storing left half of image1 in left half of image out
    image_out[:, 0: image_out.shape[1]/2] =  image1[:, 0:image1.shape[1]/2] 
    ##storing right half of image1 in right half of image out
    image_out[:, image_out.shape[1]/2:image_out.shape[1]] =  image2[:, image1.shape[1]/2:image1.shape[1]]
    return image_out
    

```

```python

def alpha(image1,image2):
   

    image1 = img_as_float(image1)
    image2 = img_as_float(image2)
    
    alpha_window = 10 ##Defining window for masking
    alpha_image = np.zeros((image1.shape[0],alpha_window*2,3))
    
    alpha_image[:,:,0] = alpha_image[:,:,1] = alpha_image[:,:,2] = \
        np.tile(np.linspace(0, 1, alpha_window*2, endpoint=True), (image1.shape[0],1))

    #First simple blending is performed
    im_out = simple(image1,image2)

    #then the masked region is filled with alpha blending 
    im_out[:, int(im_out.shape[1]/2)-alpha_window:int(im_out.shape[1]/2)+alpha_window] = \
    image1[:, int(im_out.shape[1]/2)-alpha_window:int(im_out.shape[1]/2)+alpha_window]*(1-alpha_image) + \
    image2[:, int(im_out.shape[1]/2)-alpha_window:int(im_out.shape[1]/2)+alpha_window]*alpha_image 

    return im_out
```

```python
 def pyramid(image1,image2):       
        p1 = np.matrix(np.zeros((512,256)))
        p2 = np.matrix(np.ones((512,256)))
        mask = np.concatenate((p1,p2),axis=1)
        
        gaussian_pyramid_image1 = transform.pyramid_gaussian(image1,max_layer=3)
        gaussian_pyramid_image1 = [a for a in gaussian_pyramid_image1]
        gaussian_pyramid_image2 = transform.pyramid_gaussian(image2,max_layer=3)
        gaussian_pyramid_image2 = [a for a in gaussian_pyramid_image2]
      
        laplacian_pyramid_image1 = []
        laplacian_pyramid_image2 = []
        
        for i in range(len(gaussian_pyramid_image1)-1):
            laplacian_pyramid_image1.append(gaussian_pyramid_image1[i] - transform.rescale((gaussian_pyramid_image1[i+1]),2.0))
            laplacian_pyramid_image2.append(gaussian_pyramid_image2[i] - transform.rescale((gaussian_pyramid_image2[i+1]),2.0))
            
        laplacian_pyramid_image1.append(gaussian_pyramid_image1[3])
        laplacian_pyramid_image2.append(gaussian_pyramid_image2[3])
        
        gaussian_pyramid_mask = transform.pyramid_gaussian(mask,max_layer=3)
        gaussian_pyramid_mask = [a for a in gaussian_pyramid_mask]
     
        blend_pyramid = []
        for i in range(len(gaussian_pyramid_mask)):
            blend_pyramid.append(laplacian_pyramid_im1[i]*gaussian_pyramid_mask[i] + laplacian_pyramid_image2[i]*(1-gaussian_pyramid_mask[i]))
            
        blend_pyramid_final = transform.rescale((transform.rescale((transform.rescale(blend_pyramid[3],2.0)+blend_pyramid[2]),2)+blend_pyramid[1]),2.0)+blend_pyramid[0]
        return blend_pyramid_final
```
